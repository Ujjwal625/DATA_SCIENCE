Replacing per column





```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.NaN)
new_df
```

Replacing by using mapping



```python
new_df = df.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df
```

Regex  REGULAR EXPRESSION




```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]',
                     'windspeed': '[a-z]'}, '', regex=True) 
new_df
```
