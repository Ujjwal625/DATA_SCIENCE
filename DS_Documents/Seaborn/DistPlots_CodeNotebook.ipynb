```python
# Distplots in Seaborn

# Importing the libraries
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Loading the Penguins Dataset


penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
#Drawing a Simple distplot

# here,The y axis gives the probability density function of the data

sns.distplot(penguins.flipper_length_mm,rug = True) #penguins.flipper_length_mm==COLUMN(of which chart is ploted)
```

```python
#Removing the KDE Plot


sns.distplot(penguins.flipper_length_mm,hist = False)
```

```python
#Setting the number of bins


sns.distplot(penguins.flipper_length_mm,bins = 20)
```

```python
#Styling Rugplots


sns.distplot(penguins.flipper_length_mm,rug_kws = {'color':'red','height':0.30},rug = True)
```

```python
#Styling Histograms

sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'blue','alpha':0.2,'histtype':'stepfilled'})
# hist_kws is here for costumising
```

```python
sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'blue','alpha':0.2,'histtype':'step'})
#just differ in histtype  ,, alpha is colour gradient
```

```python
#Styling KDE Plots

sns.distplot(penguins.flipper_length_mm,kde = True,hist = False,kde_kws = {'color':'red','lw':4.0,'alpha':0.9})
# lw is for line width
```
