CSV file is available to download at https://github.com/codebasics/py/blob/master/ML/9_decision_tree/Exercise/titanic.csv

In this file using following columns build a model to predict if person would survive or not,
Pclass
Sex
Age
Fare
Calculate score of your model

```python
import pandas as pd
df = pd.read_csv('F:\\Jupyter_notebook\\csv_xlxs_File\\TITANIC.csv')
df.head()
```

```python
inputs=df[['Pclass','Sex','Age','Fare']]
inputs
```

```python
target=df.Survived
```

```python
target
```

```python

from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()

inputs['Sex']=labelencoder.fit_transform(inputs['Sex'])

inputs
```

```python
inputs['Age']=df['Age'].fillna(inputs['Age'].mean())
```

```python

from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3) 
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
