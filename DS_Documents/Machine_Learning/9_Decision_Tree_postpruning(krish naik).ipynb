```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')  # we can also loat the data from  seaborn instead of sklearn 
df.head()
```

```python
#independent feature and dependent features
X=df.iloc[:,:-1]  # here -1 represent that except the last column i.e 'species' we require every thin in 'X'
y=iris.target
```

```python
X,y
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

treemodel=DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#prediction
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
#see,here score is 1 means 100%, which come because sometime decision tree leads to overfitting(which is not good) ,
#and the smae is done here.
# later in the same code you see that post prunning is done(max_depth), due to which the  score remains 98%, which is good
```

```python
print(classification_report(y_pred,y_test))
```

```python

## Postpruning
treemodel_new=DecisionTreeClassifier(max_depth=2)
#here post-pruning is done here feature which is used here is (max_depth) 
```

```python
treemodel_new.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel_new,filled=True)
```

```python
#prediction
y_pred_new=treemodel_new.predict(X_test)
```

```python
y_pred_new
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred_new,y_test)
print(score)
#see above ,there score is 100% which is due to overfitting , which is bad  for model,
#so we done postpruning here, and score 98%
```
